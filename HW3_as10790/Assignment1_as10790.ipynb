```python
from __future__ import print_function
import pylab as pl
from scipy.optimize import curve_fit, minimize
    
%pylab inline
```

```python
pop_mean = 100
sample_size = np.array(np.linspace(11, 1999, 100, dtype = int))
```

1) Normal distribution

```python
np.random.seed(111)
norm_samples = [np.random.normal(loc = pop_mean, size = N) for N in sample_size]
norm_means = [np.mean(i) for i in norm_samples]

pl.scatter(sample_size, norm_means)
pl.xlabel('Sample size'), pl.ylabel('Sample mean'), pl.title('Normal distribution')
```

2) Poisson distribution

```python
np.random.seed(122)
poi_samples = [np.random.poisson(lam = 100, size = N) for N in sample_size]
poi_means = [np.mean(i) for i in poi_samples]

pl.scatter(sample_size, poi_means)
pl.xlabel('Sample size'), pl.ylabel('Sample mean'), pl.title('Poisson distribution')
```

3) Binomial distribution

```python
np.random.seed(121)
n = 200
p = 0.5

bi_samples = [np.random.binomial(n, p, N) for N in sample_size]
bi_means = [np.mean(i) for i in bi_samples]

pl.scatter(sample_size, bi_means)
pl.xlabel('Sample size'), pl.ylabel('Sample mean'), pl.title('Binomial distribution')
```

4) Chi-square distribution

```python
np.random.seed(229)

df = pop_mean
chi_samples = [np.random.chisquare(df, size = N) for N in sample_size]
chi_means = [np.mean(i) for i in chi_samples]

pl.scatter(sample_size, chi_means)
pl.xlabel('Sample size'), pl.ylabel('Sample mean'), pl.title('Chi-square distribution')
```

5) Uniform distribution

```python
np.random.seed(100)
uni_sample = [np.random.uniform(high = 200, size = N) for N in sample_size]
uni_means = [np.mean(i) for i in uni_sample]

pl.scatter(sample_size, uni_means)
pl.xlabel('Sample size'), pl.ylabel('Sample mean'), pl.title('Uniform distribution')
```

# The plots suggest that as the sample size increases, the sample mean gets closer to the expected value, i.e. the population mean of 100. 

Central Limit Theorem

```python
allmeans = np.concatenate((norm_means, poi_means, bi_means, chi_means, uni_means))
pl.hist(allmeans, bins = 50)
pl.xlabel('Sample mean')
pl.ylabel('N')

```
