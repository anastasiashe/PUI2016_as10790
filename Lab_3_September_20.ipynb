```python
import numpy as np
import matplotlib.pylab as plt

%pylab inline
```

```python
np.random.seed(229) # This is good to do 
df = 100 # Degrees of freedom
chisq = np.random.chisquare(df, size = 100)
```

```python
plt.hist(chisq)
```

```python
chisq_mean = chisq.mean()
print chisq_mean
```

```python
# Include a docstring starting and ending with """ """  explaining what the function does
def chisq_size(df, n):
    return np.random.chisquare(df, size = n).mean()
```

```python
np.random.seed(229) # This is good to do 
df = 100 # Degrees of freedom
chisq = np.random.chisquare(df, size = 100)

# Pythonian list comprehension! 
chisq_mean = [chisq_size(df, n) for n in np.linspace(10, 1000, 100)] #linspace better than range, np.arange vs. range 
print(chisq_mean)
```

```python
sample_sizes = np.linspace(10,1000,100)
plt.figure()
plt.scatter(sample_sizes, chisq_mean)
```
