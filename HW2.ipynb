```python
import sys
import json
try:
    import urllib2 as urllib
except ImportError:
    import urlib.request as urllib

def get_jsonparsed_data(url):
    response = urllib.urlopen(url)
    data = response.read().decode("utf-8")
    return json.loads(data)

jsonData = get_jsonparsed_data("http://bustime.mta.info/api/siri/vehicle-monitoring.json?key=11b6806c-c9bc-404b-9b10-468d9765fb46&VehicleMonitoringDetailLevel=calls&LineRef=B38")
jsonData
```
